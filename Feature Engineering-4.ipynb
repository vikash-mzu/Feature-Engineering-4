```python
Q1: What is data encoding? How is it useful in data science?
Data encoding is the process of transforming categorical (non-numeric) data into a numerical format that machine learning algorithms can process. Since most ML models can only work with numerical data, encoding is essential for converting text-based features like names, categories, or labels into numbers.
•	Usefulness in data science: It allows algorithms to make sense of categorical variables. Without encoding, models such as linear regression, decision trees, or neural networks wouldn’t be able to process textual data.
________________________________________
Q2: What is nominal encoding? Provide an example of how you would use it in a real-world scenario.
Nominal encoding (also known as label encoding) assigns each unique category in a categorical feature to a specific integer value. It is useful for ordinal features, where there is no inherent ranking or relationship between categories.
Example: Suppose you are working on a project for a grocery store and have a feature FruitType with categories: ['Apple', 'Banana', 'Orange', 'Grapes']. Using nominal encoding:
•	Apple = 0
•	Banana = 1
•	Orange = 2
•	Grapes = 3
This would transform the categorical data into numbers, making it suitable for machine learning models.
________________________________________
Q3: In what situations is nominal encoding preferred over one-hot encoding? Provide a practical example.
Nominal encoding is preferred over one-hot encoding when:
•	There are a large number of categories.
•	The model being used can naturally handle integer-encoded data (e.g., decision trees).
•	The categorical feature has no meaningful order, and we don’t need to prevent the model from imposing an ordinal relationship between the categories.
Example: In a recommendation system for an e-commerce platform, if you have a categorical feature representing 100 different product categories, using one-hot encoding would create 100 new binary columns. Nominal encoding, on the other hand, would assign an integer to each product category and only create a single column, thus saving memory and reducing complexity.
________________________________________
Q4: Suppose you have a dataset containing categorical data with 5 unique values. Which encoding technique would you use to transform this data into a format suitable for machine learning algorithms? Explain why you made this choice.
If the categorical data has no inherent order and only 5 unique values, I would use one-hot encoding. This is because:
•	One-hot encoding creates binary columns for each category, ensuring that the model doesn’t infer any ordinal relationships between categories that don’t exist.
•	With only 5 unique values, one-hot encoding won’t introduce too many additional columns.
For example, if the categories are ['Red', 'Blue', 'Green', 'Yellow', 'Purple'], one-hot encoding would create 5 binary columns, where each column corresponds to one of the colors.
________________________________________
Q5: In a machine learning project, you have a dataset with 1000 rows and 5 columns. Two of the columns are categorical, and the remaining three columns are numerical. If you were to use nominal encoding to transform the categorical data, how many new columns would be created? Show your calculations.
In nominal encoding, each categorical column is replaced with a single column of integers representing the different categories.
•	Assumption: Let’s assume the first categorical column has mmm unique values, and the second has nnn unique values.
Since nominal encoding assigns a single integer to each category, the number of new columns created would still be 2 (one for each categorical feature). No extra columns are added in nominal encoding, unlike one-hot encoding.
________________________________________
Q6: You are working with a dataset containing information about different types of animals, including their species, habitat, and diet. Which encoding technique would you use to transform the categorical data into a format suitable for machine learning algorithms? Justify your answer.
For this dataset containing features like species, habitat, and diet, I would use one-hot encoding. This is because:
1.	These features are likely nominal, meaning there is no intrinsic order to species or habitat.
2.	One-hot encoding prevents the model from assuming any ordinal relationship between categories, which is crucial for nominal variables.
3.	The number of unique categories in each feature is likely manageable, and one-hot encoding would allow the model to treat each category independently.
For example:
•	Species: ['Mammal', 'Bird', 'Fish']
•	Habitat: ['Forest', 'Desert', 'Ocean']
One-hot encoding would create binary columns for each unique category.
________________________________________
Q7: You are working on a project that involves predicting customer churn for a telecommunications company. You have a dataset with 5 features, including the customer's gender, age, contract type, monthly charges, and tenure. Which encoding technique(s) would you use to transform the categorical data into numerical data? Provide a step-by-step explanation of how you would implement the encoding.
In this scenario, the categorical features are likely gender and contract type, while age, monthly charges, and tenure are numerical.
Step-by-step encoding:
1.	Identify categorical features:
o	Gender: Typically has 2 values (e.g., Male, Female).
o	Contract Type: May have 3 values (e.g., Month-to-month, One-year, Two-year).
2.	Apply encoding techniques:
o	For Gender: Since it is a binary feature (Male/Female), you could use binary encoding or one-hot encoding. In this case, one-hot encoding would create two columns: Gender_Male and Gender_Female.
o	For Contract Type: Since it has 3 unique values, apply one-hot encoding. This would create 3 columns: Contract_Month-to-month, Contract_One-year, and Contract_Two-year.
3.	Implement encoding:
o	Using one-hot encoding:
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

data = pd.DataFrame({
    'gender': ['Male', 'Female', 'Male'],
    'contract_type': ['Month-to-month', 'One-year', 'Two-year']
})

encoder = OneHotEncoder(drop='first')  # drop='first' to avoid multicollinearity
encoded_data = pd.DataFrame(encoder.fit_transform(data[['gender', 'contract_type']]).toarray())
4.	Result: The dataset will now include the one-hot encoded columns for gender and contract type, while the numerical columns (age, monthly charges, tenure) remain unchanged.

```
